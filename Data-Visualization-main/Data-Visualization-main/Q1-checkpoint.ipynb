```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset("penguins")

# KDE plot for multiple variables
plt.figure(figsize=(12, 6))
sns.kdeplot(data=data, x="bill_length_mm", hue="species", fill=True, alpha=0.6, palette="muted")
plt.title("Kernel Density Estimation of Bill Length by Species", fontsize=16)
plt.xlabel("Bill Length (mm)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
