```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset("tips")

# Bivariate scatter plot with regression line
sns.jointplot(data=data, x="total_bill", y="tip", kind="reg", height=8)
plt.suptitle("Bivariate Distribution of Total Bill vs Tip", y=1.02, fontsize=16)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset("tips")

# Bivariate scatter plot with hue
sns.jointplot(data=data, x="total_bill", y="tip", hue="sex", kind="scatter", height=8)
plt.suptitle("Bivariate Distribution of Total Bill vs Tip by Gender", y=1.02, fontsize=16)
plt.show()
```
