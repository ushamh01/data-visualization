```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Create a graph
G = nx.Graph()

# Add nodes with coordinates
for i in range(10):
    G.add_node(i, pos=(random.uniform(0, 10), random.uniform(0, 10)))

# Add random edges
edges = [(i, random.choice(list(G.nodes))) for i in G.nodes]
G.add_edges_from(edges)

# Draw the network
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="gray", font_size=10)
plt.title("Network Graph with Random Geospatial Connections", fontsize=16)
plt.show()
```
